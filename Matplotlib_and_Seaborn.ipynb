### 2D Line plot

image.png

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price, marker='o', linestyle='dashed',linewidth=2, markersize=12)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['RG Sharma'], marker='o',linewidth=2, markersize=12,label='RG Sharma',color='blue')
plt.plot(batsman['index'],batsman['V Kohli'], marker='o',linewidth=2, markersize=12,label='V Kohli',color='red')

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# solid,dotted,dashed,dashdot
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle="dashdot",linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='solid',linewidth=3)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',marker='o',markersize=12)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### label and legend (To print label we have to call legend)

```python

plt.plot(batsman['index'],batsman['RG Sharma'], marker='D',linewidth=2, markersize=12,color='blue',label='RG Sharma')
plt.plot(batsman['index'],batsman['V Kohli'], marker='o',linewidth=2, markersize=12,color='red',label='V Kohli')

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel('Runs Scored')

# legend -> location
plt.legend(loc = 'upper center') # default loc = best
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2016,2019)
```

```python
# grid


plt.plot(batsman['index'],batsman['RG Sharma'], marker='D',linewidth=2, markersize=12,color='blue',label='RG Sharma')
plt.plot(batsman['index'],batsman['V Kohli'], marker='o',linewidth=2, markersize=12,color='red',label='V Kohli')

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel('Runs Scored')

# legend -> location
plt.legend(loc = 'upper center') # default loc = best
plt.grid()
```

### Show - is the main function to display the plots ,on pycharm,flask,streamlit

```python
# show
plt.plot(batsman['index'],batsman['RG Sharma'], marker='D',linewidth=2, markersize=12,color='blue',label='RG Sharma')
plt.plot(batsman['index'],batsman['V Kohli'], marker='o',linewidth=2, markersize=12,color='red',label='V Kohli')

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Seasons")
plt.ylabel('Runs Scored')

# legend -> location
plt.legend(loc = 'upper center') # default loc = best
plt.grid()
plt.show()
```

# Scatter Plots

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('/content/batter.csv')
df = df.head(50)
df.sort_values("avg",ascending=False)
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker="*")
plt.title("Avg and Strike Rate analysis of top 50 cricketer")
plt.xlabel('Avg')
plt.ylabel('Strike Rate')
```

```python
# size
tips = sns.load_dataset('tips')
tips
```

```python
# slower
plt.scatter(tips['total_bill'],tips['tip'],s = tips['size']*5)
```

```python
# scatterplot using plt.plot
# faster
plt.plot(tips['total_bill'],tips['tip'],'o')
```

```python
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

## Multiple bar plot

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='orange',label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red',label='2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='blue',label='2017')

plt.legend()
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
# xticks
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ["Tea", "Coffee"]
male_counts = [40, 60]
female_counts = [70, 30]

x = np.arange(len(categories))
width = 0.35

plt.bar(x - width/2, male_counts, width, label="Male")
plt.bar(x + width/2, female_counts, width, label="Female")

plt.xticks(x, categories)
plt.xlabel("Drink Preference (Category)")
plt.ylabel("Count")
plt.title("Gender vs Drink Preference")
plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

# percentage = autopct = "%0.1f%%

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
