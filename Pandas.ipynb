````python
batsman['batsman_run'].rank(ascending=False, method='average')  # default
batsman['batsman_run'].rank(ascending=False, method='min')      # lowest rank in tie
batsman['batsman_run'].rank(ascending=False, method='max')      # highest rank in tie
batsman['batsman_run'].rank(ascending=False, method='first')    # order of appearance
batsman['batsman_run'].rank(ascending=False, method='dense')
````

For [50, 30, 50, 10]:

average → [1.5, 3.0, 1.5, 4.0]

min → [1.0, 3.0, 1.0, 4.0]

max → [2.0, 3.0, 2.0, 4.0]

first → [1.0, 3.0, 2.0, 4.0]

dense → [1.0, 2.0, 1.0, 3.0]

````python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
````

````python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
````

````python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
````
